# Grid Axis and labels

```python
x = np.arange(3)
plt.plot(x,x**2, x,x**3, x,x*2, x,2**x)
plt.grid(True) #shows the grid lines
print(plt.axis()) 
```

```python
plt.plot(x,x**2, x,x**3, x,x*2, x,2**x)
plt.grid(True)
plt.axis([0,2,0,8]) #modify where ther graph begins and ends
```

```python
plt.plot(x,x**2, x,x**3, x,x*2, x,2**x)
plt.grid(True)
plt.xlim([0,2])
plt.ylim([0,8]) #another way to define graph limits
```

```python
plt.plot(x,x**2, x,x**3, x,x*2, x,2**x)
plt.grid(True)
plt.xlabel('x = np.arange(3)') #x label
plt.ylabel('y = f(x)')          #y label
plt.xlim([0,2])
plt.ylim([0,8])
plt. title('Graph Title here')
```

```python
plt.plot(x,x**2, label = 'x^2')
plt.plot(x,x**3, label = 'x^3')
plt.plot(x,x*2, label = '2x')
plt.plot(x,2**x, label = '2^x')
plt.legend()
plt.grid(True)
plt.xlabel('x = np.arange(3)') #x label
plt.ylabel('y = f(x)')          #y label
plt.xlim([0,2])
plt.ylim([0,8])
plt. title('Graph Title here')
```

```python
plt.plot(x,x**2, x,x**3, x,x*2, x,2**x)
plt.legend(['x^2','x^3','2x','2^x'], loc='upper center')
plt.grid(True)
plt.xlabel('x = np.arange(3)') #x label
plt.ylabel('y = f(x)')          #y label
plt.xlim([0,2])
plt.ylim([0,8])
plt. title('Graph Title here')
```

```python
plt.plot(x,x**2, x,x**3, x,x*2, x,2**x)
plt.legend(['x^2','x^3','2x','2^x'], loc='upper center')
plt.grid(True)
plt.xlabel('x = np.arange(3)') #x label
plt.ylabel('y = f(x)')          #y label
plt.xlim([0,2])
plt.ylim([0,8])
plt. title('Graph Title here')
plt.savefig('test.png')
```

# color and line markers

```python
plt.plot(x,x**2, 'g', label = 'x^2')
plt.plot(x,x**3, 'y', label = 'x^3')
plt.plot(x,x*2,'r', label = '2x')
plt.plot(x,2**x,'k', label = '2^x')
plt.legend()
plt.grid(True)
plt.xlabel('x = np.arange(3)') #x label
plt.ylabel('y = f(x)')          #y label
plt.xlim([0,2])
plt.ylim([0,8])
plt. title('Graph Title here')
```

```python
plt.plot(x,x**2, ':', label = 'x^2')
plt.plot(x,x**3, '--',  label = 'x^3')
plt.plot(x,x*2,'-',  label = '2x')
plt.plot(x,2**x,'-.',  label = '2^x')
plt.legend()
plt.grid(True)
plt.xlabel('x = np.arange(3)') #x label
plt.ylabel('y = f(x)')          #y label
plt.xlim([0,2])
plt.ylim([0,8])
plt. title('Graph Title here')
```

```python
plt.plot(x,x**2, 'ro:', label = 'x^2')
plt.plot(x,x**3, 'g<--',  label = 'x^3')
plt.plot(x,x*2,'y*-.',  label = '2x')
plt.plot(x,2**x,'m^-',  label = '2^x')
plt.legend()
plt.grid(True)
plt.xlabel('x = np.arange(3)') #x label
plt.ylabel('y = f(x)')          #y label
plt.xlim([0,2])
plt.ylim([0,8])
plt. title('Graph Title here')
```
